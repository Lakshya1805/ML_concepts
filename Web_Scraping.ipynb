```python
final = pd.DataFrame()
for j in range(1,9):
    url = 'https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_1_books?ie=UTF8&pg={}'.format(j)
    webpage = requests.get(url).text
    soup = BeautifulSoup(webpage,'lxml')
    
    book = soup.find_all('div',class_='p13n-sc-uncoverable-faceout')
    
    book_name = []
    author = []
    rating = []
    total_reviews = []
    price = []
    book_type = []


    for i in book:
        div_elements = i.find_all('div', class_='_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y')
        if len(div_elements) >= 2:
            book_name.append(div_elements[0].text)
            author.append(div_elements[1].text)
        else:
            book_name.append("N/A")
            author.append("N/A")

        span_element = i.find('span', class_='a-icon-alt')
        if span_element is not None:
            rating.append(span_element.text)
        else:

            rating.append("N/A")

        price.append(i.find('span',class_='p13n-sc-price').text)

        tag_elements = i.find_all('span',class_='a-size-small')
        if len(tag_elements) == 2:
            total_reviews.append(tag_elements[0].text)
            book_type.append(tag_elements[1].text)
        else:
            total_reviews.append(tag_elements[1].text)
            book_type.append(tag_elements[2].text)
            
        dict = {'Name':book_name, 'Author':author, 'Rating':rating,'Number of Reviews':total_reviews,'Price':price,'Type':book_type}
        df = pd.DataFrame(dict)
        final = final.append(df, ignore_index=True)
    
    
    
```
