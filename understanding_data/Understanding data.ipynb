# Using Column Transformer

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers = [('tnf1',SimpleImputer(),['Age']),
                                                ('tnf2', OneHotEncoder(sparse=False, drop = 'first' ), ['Sex','Embarked'])], remainder = 'passthrough')
```

```python
x_train_done = transformer.fit_transform(x_train)
```

```python
x_test_done = transformer.transform(x_test)
```
