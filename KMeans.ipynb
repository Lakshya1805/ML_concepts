```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs
```

```python
x,y = make_blobs(n_samples = 500, n_features =2,centers=8)
```

```python
x
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
from sklearn.cluster import KMeans
```

```python
wcs = []
for i in range(1,20):
    km = KMeans(n_clusters=i)
    km.fit(x)
    wcs.append(km.inertia_)
```

```python
wcs
```

```python
plt.plot(wcs)
```

```python
km = KMeans(n_clusters=7)
```

```python
y_pred = km.fit_predict(x)
```

```python
df = pd.DataFrame()
```

```python
df['col1'] = x[:,0]
df['col2'] = x[:,1]
df['label'] = y_pred
```

```python
df
```

```python
plt.scatter(df['col1'],df['col2'],c = df['label'])
```
