```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import make_moons,make_blobs,make_circles
```

```python
x1,y1 = make_circles(n_samples = 500,noise =0.05,factor=0.5)
```

```python
x2,y2 = make_moons(n_samples=500,noise=0.05)
x3,y3 = make_blobs(n_samples=500,n_features=2,centers=4,)
```

```python
import matplotlib.pyplot as plt
```

```python
y1.shape
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Scatter plot for the first data
axes[0].scatter(x1[:, 0], x1[:, 1])
axes[0].set_title("Plot 1")

# Scatter plot for the second data
axes[1].scatter(x2[:, 0], x2[:, 1])
axes[1].set_title("Plot 2")

# Scatter plot for the third data
axes[2].scatter(x3[:, 0], x3[:, 1])
axes[2].set_title("Plot 3")

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
import scipy.cluster.hierarchy as shc
```

```python
dend1 = shc.dendrogram(shc.linkage(x1,method='ward'))
```

```python
dend2 = shc.dendrogram(shc.linkage(x2,method='ward'))
```

```python
dend3 = shc.dendrogram(shc.linkage(x3,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac1 = AgglomerativeClustering(n_clusters=2,linkage='single',metric='euclidean')
```

```python
x1_pred = ac1.fit_predict(x1)
```

```python
ac2 = AgglomerativeClustering(n_clusters=2,linkage='single',metric='euclidean')
x2_pred = ac2.fit_predict(x2)
```

```python
ac3 = AgglomerativeClustering(n_clusters = 4,linkage='single',metric='euclidean')
x3_pred = ac3.fit_predict(x3)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Scatter plot for the first data
axes[0].scatter(x1[:, 0], x1[:, 1],c=x1_pred)
axes[0].set_title("Plot 1")

# Scatter plot for the second data
axes[1].scatter(x2[:, 0], x2[:, 1],c=x2_pred)
axes[1].set_title("Plot 2")

# Scatter plot for the third data
axes[2].scatter(x3[:, 0], x3[:, 1],c=x3_pred)
axes[2].set_title("Plot 3")

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
